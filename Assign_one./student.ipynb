```python
import pandas as pd
import numpy as np
```

```python
student =pd.read_excel("student.xlsx")
student
```

```python
student.isnull().sum()
```

```python
#using backfill to compute for the missing values
student.fillna(method='bfill',inplace=True)
student
```

```python
#using forward fill to compute for the missing values
student.fillna(method='ffill',inplace=True)
student
```

```python
#using interpolation to compute for the missing values
student.interpolate(method='linear',inplace=True)
student
```

```python
#seperating my dataset into cont_data
cont_data =student.select_dtypes(include=['int64','float64'])
cont_data

```

```python
#separating my data into cat_data
cat_data =student.select_dtypes(include=['object'])
cat_data
```
