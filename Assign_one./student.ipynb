```python
import pandas as pd
import numpy as np
```

```python
student =pd.read_excel("student.xlsx")
student
```

```python
student.isnull().sum()
```

```python
#using backfill to compute for the missing values
student.fillna(method='bfill',inplace=True)
student
```

```python
student.isnull().sum()
```

```python
#using forward fill to compute for the missing values
student.fillna(method='ffill',inplace=True)
student
```

```python
student.isnull().sum()
```

```python
#using interpolation to compute for the missing values
student.interpolate(method='linear',inplace=True)
student
```

```python
student.isnull().sum()
```

```python
#seperating my dataset into cont_data
cont_data =student.select_dtypes(include=['int64','float64'])
cont_data

```

```python
#separating my data into cat_data
cat_data =student.select_dtypes(include=['object'])
cat_data
```

```python
#visualizing to know  which methode to use for imputing

import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
for i, col in enumerate(cont_data.columns):
    plt.subplot(3, 3, i+1)
    plt.hist(student[col])
    plt.title(col)
    plt.ylabel(f"frequency")
plt.tight_layout()
plt.show()
```

```python
#filling in for the missing values using mean
#we use the mean for continous data
student['DSC3108'].fillna(student['DSC3108'].mean(),inplace=True)
student['DSC3114'].fillna(student['DSC3114'].mean(),inplace=True)
student['MTH3108'].fillna(student['MTH3108'].mean(),inplace=True)
student['SYE3104'].fillna(student['SYE3104'].mean(),inplace=True)
```

```python
#we use mode for categorical data
student['course'].fillna(student['course'].mode()[0],inplace=True)
```
